```python
import numpy as np
import seaborn as sns; sns.set_theme()
sns.set(font_scale=1)
```

```python
#Funci칩n con par치metros
def plot_matrix(matriz, classes, title):
  ax = sns.heatmap(matriz, cmap="Blues", annot=True, xticklabels=classes, yticklabels=classes, cbar=True)
  ax.set(title=title, xlabel="Predicted label", ylabel="Actual label")
```

```python
matriz = np.array([[87, 16], [17, 59]])
classes = ['0', '1']
title = "Matriz de Confusi칩n"

plot_matrix(matriz, classes, title)

#Fuente: https://stackoverflow.com/questions/60480777/plotting-already-calculated-confusion-matrix-using-python
```
